### Implement a circuit that returns $\lvert01\rangle$ and $\lvert10\rangle$ with equal probability (50% for each).

Using H, X, CX gates we can make this circuit as following:

```python
q = QuantumCircuit(2,2)

# q.x(0)
q.x(1)
q.h(0)
q.cx(0,1)
q.barrier(range(2))
q.measure((0,1),(0,1))

q.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(q, backend, shots= 1024)
result = job.result()
count = result.get_counts()
print(count)
plot_histogram(count)
```

#### The circuit should consist only of CNOTs, RXs and RYs.

So using Rx, Ry gates with 'fixed' parameters we can translate the previous circuit:

```python
qc = QuantumCircuit(2,2)

# qc.rx(np.pi,0)
qc.rx(np.pi,1)
qc.ry((-np.pi/2),0)
qc.rx(np.pi,0)
qc.cx(0,1)
qc.barrier(range(2))
qc.measure((0,1),(0,1))

qc.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
jobs = execute(qc, backend, shots= 1024)
results = jobs.result()
counts = results.get_counts()
print(counts)
plot_histogram(counts)
```

**But we have to find the right set of parameters being equal to 0 or randomly chosen using gradient descent.**

```python
from qiskit.aqua.components.optimizers.adam_amsgrad import ADAM
```

```python
backend_qasm = Aer.get_backend('qasm_simulator')
backend_statevec = Aer.get_backend('statevector_simulator')
```

```python
def circuit(params):
    circ = QuantumCircuit(2,2)
    
    circ.rx(params[0],1)
    circ.ry(params[1],0)
    circ.rx(params[2],0)
    circ.cx(0,1)
    
    circ.barrier(range(2))
    circ.measure((0,1),(0,1))
    
    return circ
    
```

```python
def counts(backend, circ, shot):
    job = execute(circ, backend, shots=shot)
    result = job.result()
    count = result.get_counts()
    
    return count
```

```python
def probabilities(count,shots):
    
    if '00' in count.keys():
        p_00 = count['00']/shots*100
    else:
        p_00 = 0
    if '01' in count.keys():
        p_01 = count['01']/shots*100
    else:
        p_01 = 0
    if '10' in count.keys():
        p_10 = count['10']/shots*100
    else:
        p_10 = 0
    if '11' in count.keys():
        p_11 = count['11']/shots*100
    else:
        p_11 = 0
        
    costs = ((0-p_00)**2 + (50-p_01)**2 + (50-p_10)**2 + (0-p_11)**2)
        
        
    return costs
```

```python
def cost_1(params):
    cir = circuit(params)
    count = counts(backend_qasm,cir,1)
    costs = probabilities(count,1)
    
    return costs

def cost_10(params):
    cir = circuit(params)
    count = counts(backend_qasm,cir,10)
    costs = probabilities(count,10)
    
    return costs

def cost_100(params):
    cir = circuit(params)
    count = counts(backend_qasm,cir,100)
    costs = probabilities(count,100)
    
    return costs

def cost_1000(params):
    cir = circuit(params)
    count = counts(backend_qasm,cir,1000)
    costs = probabilities(count,1000)
    
    return costs
```

```python
opt = ADAM(maxiter=1000, tol=0.01, lr=0.5, beta_1=0.9, beta_2=0.999)
steps = 5

points_costs_adam_optimizer_1 = []
points_costs_adam_optimizer_10 = []
points_costs_adam_optimizer_100 = []
points_costs_adam_optimizer_1000 = []

#for shot = 1
for i in range(steps):
    initial_point = np.random.rand(3)
    params = opt.optimize(num_vars=3, objective_function=cost_1, initial_point=initial_point)
    print(params)
    points_costs_adam_optimizer_1.append(params)
    if params[1] <= 1.00:
        break
        
        
#for shot = 10
for i in range(steps):
    initial_point = np.random.rand(3)
    params = opt.optimize(num_vars=3, objective_function=cost_10, initial_point=initial_point)
    print(params)
    points_costs_adam_optimizer_10.append(params)
    if params[1] <= 1.00:
        break
        

#for shot = 100
for i in range(steps):
    initial_point = np.random.rand(3)
    params = opt.optimize(num_vars=3, objective_function=cost_100, initial_point=initial_point)
    print(params)
    points_costs_adam_optimizer_100.append(params)
    if params[1] <= 1.00:
        break
        
#for shot = 1000
for i in range(steps):
    initial_point = np.random.rand(3)
    params = opt.optimize(num_vars=3, objective_function=cost_1000, initial_point=initial_point)
    print(params)
    points_costs_adam_optimizer_1000.append(params)
    if params[1] <= 1.00:
        break
```

```python
params = [np.pi, -np.pi/2, np.pi]
c = circuit(params)
c.draw('mpl')
shots=[1,10,100,1000]
for shot in shots:
    count = counts(backend_qasm,c,shot)
    print(count)
    costs = probabilities(count,params[-1])
    print(costs)
```
